```python
import numpy as np
import matplotlib.pyplot as plt

# 创建同位圆
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

# 画出同位圆
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Unit Circle', color='blue')

# 画出角度 pi/6
angle = np.pi / 6
plt.plot([0, np.cos(angle)], [0, np.sin(angle)], color='red', label='Angle π/6')

# 画出 cos(π/6)
plt.plot([0, np.cos(angle)], [0, 0], color='green', linestyle='--', label='cos(π/6)')

# 添加垂线段
plt.plot([np.cos(angle), np.cos(angle)], [0, np.sin(angle)], color='purple', linestyle=':', label='Vertical Line')

# 标记 cos(π/6)
plt.scatter([np.cos(angle)], [0], color='black')
plt.text(np.cos(angle), 0.1, 'cos(π/6)', horizontalalignment='center')

# 设置坐标轴
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

# 添加图例和标题
plt.title('Unit Circle and cos(π/6)')
plt.legend()
plt.grid()
plt.show()
```
