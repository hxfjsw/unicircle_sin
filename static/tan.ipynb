```python
#使用同位圆 绘制sin(π/6)的图像
import numpy as np
import matplotlib.pyplot as plt

# 创建同位圆
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

# 画出同位圆
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Unit Circle', color='blue')

# 画出角度 pi/6
angle = np.pi / 6
tan_value = np.tan(angle)
plt.plot([0, 1], [0, tan_value], color='red', label='Angle π/6')

# 添加垂线段
plt.plot([np.cos(angle), np.cos(angle)], [0, np.sin(angle)], color='purple', linestyle=':', label='Vertical Line')

# 画出 tan(π/6)
plt.plot([1, 1], [tan_value, 0], color='orange', label='tan(π/6)')
# 设置坐标轴
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.gca().set_aspect('equal', adjustable='box')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

plt.legend()
plt.grid()
plt.show()
```
