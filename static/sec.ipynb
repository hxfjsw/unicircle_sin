```python
#使用同位圆 绘制sin(π/6)的图像
import numpy as np
import matplotlib.pyplot as plt

# 创建同位圆
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

# 画出同位圆
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Unit Circle', color='blue')

# 画出角度 pi/6
angle = np.pi / 6
plt.plot([0, np.cos(angle)], [0, np.sin(angle)], color='red', label='Angle π/6')

# 画出 sec(π/6)
sec_value = 1 / np.cos(angle)
plt.plot([0, sec_value], [0, 0], color='green', label='sec(π/6)')


# 设置坐标轴
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

plt.legend()
plt.grid()
plt.show()
```
